```python
import matplotlib.pyplot as plt
import seaborn as sns
N =80 # data number
dt =15 # data step [s]
t = np.arange(0, N*dt, dt) # time
freq = np.linspace(0, 1.0/dt, N) # frequency step
close=get_mdata(N,'AUD_SGD')['close']
#np.array(close)
plt.plot(t/60,close)
```

```python
plt.plot(t/60,close-close[0])
```

```python
hum=np.hamming(N)
plt.plot(t/60,hum)
```

```python
CloseHum=(close-close[0])*np.hamming(N)
plt.plot(t/60,CloseHum)
```

```python
plt.xlim(0,0.035)
plt.ylim(0,0.00075)
plt.plot(freq,np.abs(np.fft.fft(CloseHum)))
```

```python
plt.xlim(0,0.035)
plt.ylim(0,0.00075)
plt.plot(freq,np.abs(np.fft.fft(close)))
```
